```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the datasets
dataset = pd.read_csv('merged_dataset_04.04.csv')
```

```python
dataset.head()
```

```python
# The cumulative precipitation is calculated as the sum of the daily precipitation values over time. The formula for cumulative precipitation is:
# Q_cum(n) = Q_cum(n-1) + Q(n)
# Cumulative precipitation on Day n = Cumulative precipitation on Day n-1 + precipitation on day n 
dataset['precipitation_cum'] = dataset['precipitation_mm'].cumsum()
# Reorder the columns to place 'discharge_cum' after 'discharge_mean'
columns = dataset.columns.tolist()
columns.insert(columns.index('precipitation_mm') + 1, columns.pop(columns.index('precipitation_mm')))

dataset = dataset[columns]
print(dataset)
```

```python
# The cumulative discharge is calculated as the sum of the daily discharge values over time. The formula for cumulative discharge is:
# Q_cum(n) = Q_cum(n-1) + Q(n)
# Cumulative Cischarge on Day n = Cumulative Discharge on Day n-1 + Discharge on day n 
dataset['discharge_cum'] = dataset['discharge'].cumsum()
# Reorder the columns to place 'discharge_cum' after 'discharge_mean'
columns = dataset.columns.tolist()
columns.insert(columns.index('discharge') + 1, columns.pop(columns.index('discharge')))

dataset = dataset[columns]
print(dataset)
```

```python
# Adding a new column "delta_discharge" to calculate the difference in discharge_cum
dataset['delta_discharge'] = dataset['discharge_cum'].diff() 
dataset['delta_discharge'] = dataset['delta_discharge'].fillna(0) 
```

```python
# Adding a new column "delta_discharge" to calculate the difference in discharge_cum
dataset['delta_precipitation'] = dataset['precipitation_cum'].diff() 
dataset['delta_precipitation'] = dataset['delta_precipitation'].fillna(0) 
```

```python
#drop the columns that are not needed
hydro_dataset = dataset.drop(['discharge_max', 'discharge_min', 'water_level_max', 'water_level_min', 'om_discharge', 'soil_moisture_0_to_100cm_mean', 'soil_moisture_0_to_7cm_mean', 'rain_sum', 'weather_code', 'weather_description'], axis=1)
```
